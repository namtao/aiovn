# Ensemble Learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
dataset_path = './Housing.csv'
df = pd.read_csv(dataset_path)
df
```

```python
# Xác định các cột không phải số để thực hiện chuyển
categorical_cols = df.select_dtypes(include=['object']).columns.to_list()
categorical_cols
```

```python
# Mã hoá các cột sang số
ordinal_encoder = OrdinalEncoder()
encoded_categorical_cols = ordinal_encoder.fit_transform(df[categorical_cols])

# Chuyển đổi các cột mã hoá sang df
encoded_categorical_df = pd.DataFrame(
    encoded_categorical_cols, columns=categorical_cols
)

# Loại bỏ các cột loại gốc và giữ lại các cột mã hoá 
numerical_df = df.drop(categorical_cols, axis=1)

# Gộp các cột số với các cột đã mã hoá và tạo df mới 
encoded_df = pd.concat([numerical_df, encoded_categorical_df], axis=1)
encoded_df
```

```python
# Chuẩn hoá bộ dữ liệu
normalizer = StandardScaler()
dataset_arr = normalizer.fit_transform(encoded_df)
dataset_arr
```

```python
# Tách dữ liệu
X, y = dataset_arr[:, 1:], dataset_arr[:, 0]
```

```python
# Chia dữ liệu thành train và test
test_size = 0.3
random_state = 1
is_shuffle = True
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=test_size, random_state=random_state, shuffle=is_shuffle
)
```

```python
# Train model sử dụng RandomForestRegressor
regressor = RandomForestRegressor(random_state=random_state)
regressor.fit(X_train, y_train)
```

```python
# Train model sử dụng AdaBoostRegressor
regressor = AdaBoostRegressor(random_state=random_state)
regressor.fit(X_train, y_train)
```

```python
# Train model sử dụng GradientBoostingRegressor
regressor = GradientBoostingRegressor(random_state=random_state)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_val)
```

```python
mae = mean_absolute_error(y_val, y_pred)
mse = mean_squared_error(y_val, y_pred)

print("Evaluation results on validation set:")
print(f"Mean Absolute Error: { mae}")
print(f"Mean Squared Error: { mse}")
```
