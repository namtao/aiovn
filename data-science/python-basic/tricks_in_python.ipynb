## Walrus Operator

Toán tử gán trong python ':='

```python
numbers = [2, 8, 0, 1, 1, 9, 7, 7]

description = {
    "length": (num_length := len(numbers)),
    "sum": (num_sum := sum(numbers)),
    "mean": (num_mean := num_sum / num_length),
}

print(num_length)
print(num_sum)
print(num_mean)
print(description)
```

## String methods

Có thể sử dụng '_' để phân tách hàng của số để dễ đọc

```python
n: int = 1_000_000_000
print(f"{n}")
```

Sử dụng f-string để biểu diễn số đẹp hơn để phân tách các hàng của số


```python
n: int = 1_000_000_000
print(f"{n:_}")
print(f"{n:,}")
```

Sử dụng '<', '>' hay '^' sau ':' của giá trị để biểu diễn chuỗi

```python
s: str = 'I Love U'
print(f"{s:|<20}")
print(f"{s:_>20}")
print(f"{s:#^20}")
```

Sử dụng f-string để biểu diễn các số lượng chữ số sau dấy '.' của float (Kết hợp với ',' để phân tách các hàng của số)

```python
f: float = 123123.456
print(f"{f:,.2f}")
```

Sử dụng f-string để in kết quả của phép toán bằng cách thêm dấu '=' vào phép toán

```python
a: int = 5
b: int = 3

print(f"{a + b = }")
```

Sử dụng `center` để căn giữa string 

```python
text = "I Love U"

print(text.center(30, '.'))
```

Sử dụng `ljust` `rjust` để thêm ký tự bên phải, trái của chuỗi

```python
text = "I Love U"

print(text.ljust(20, '.'))
print(text.rjust(20, '.'))
```

## Set operations

Các phép toán trên set

```python
set_a: set[int] = {1, 2, 3, 4, 5}
set_b: set[int] = {4, 5, 6, 7, 8}
```

Phép OR

```python
print(set_a | set_b)
```

Phép trừ

```python
print(set_a - set_b)
print(set_b - set_a)
```

Phép AND

```python
print(set_a & set_b)
```

Phép XOR

```python
print(set_a ^ set_b)
```
